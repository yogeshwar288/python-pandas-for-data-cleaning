```python
import pandas as pd
```

```python
df1=pd.DataFrame({'id':[1,2,3,4],'class':[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({'id':[1,2,3,4],'name':['a','b','c','d']})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='id')
```

```python
pd.merge(df2,df1,on='id')
```

```python
df3=pd.DataFrame({'id':[1,2,3,5],'name':['a','b','c','d']})
df3
```

```python
pd.merge(df3,df1,on='id')
```

```python
pd.merge(df3,df1,on='id',how='inner')
```

```python
pd.merge(df3,df1,on='id',how='left')
```

```python
pd.merge(df3,df1,on='id',how='right')
```

```python
pd.merge(df1,df3,on='id',how='outer')
```

```python
pd.merge(df3,df1,on='id',how='outer',indicator=True)
```
